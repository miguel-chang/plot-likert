```python
import plot_likert
```

```python
import numpy as np
import pandas as pd
```

```python
# Scale based on agree5 with a trailing "N/A" option
agree_with_na = plot_likert.scales.agree + ["N/A"]
print(agree_with_na)
```

```python
# Data with some NaN values (respondents who skipped the question)
np.random.seed(42)
data_with_na = pd.DataFrame(
    np.random.choice(plot_likert.scales.agree, (15, 3)),
    columns=["Question 1", "Question 2", "Question 3"],
)
# Introduce NaN for respondents who selected "N/A" / skipped
data_with_na.loc[[1, 4, 7], "Question 1"] = float("nan")
data_with_na.loc[[2, 9], "Question 2"] = float("nan")
data_with_na.loc[[0, 3, 5, 11], "Question 3"] = float("nan")
# Replace remaining responses with the matching agree_with_na labels
# (scale entries are identical; only NaN rows differ)
data_with_na
```

```python
plot_likert.plot_likert(data_with_na,plot_likert.scales.agree,plot_percentage=True,plot_nan=False,figsize=(12, 3))

plot_likert.plot_likert(data_with_na,agree_with_na,plot_percentage=True,plot_nan=False,figsize=(12, 3))
```

Passing a custom defined scale with "N/A" as an explicit extra category yields the same chart, as expected.

### Display subplot with N/A values 

```python
plot_likert.plot_likert(data_with_na,agree_with_na,plot_percentage=True,plot_nan=True,figsize=(12, 3))
```

### Display plot with centering disabled

```python
plot_likert.plot_likert(
    data_with_na,
    agree_with_na,
    plot_percentage=True,
    plot_nan=True,
    plot_center_line=False,
    figsize=(12, 3),
);
```

```python
plot_likert.plot_likert(
    data_with_na,
    agree_with_na,
    plot_percentage=True,
    plot_center_line=False,
    figsize=(12, 3),
);
```
